```python
import pandas as pd
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```python
spam = pd.read_csv("spam.csv", encoding='iso-8859-1')
```

```python
spam
```

```python
spam.groupby('v1').describe()
```

```python
spam["spam"]=spam['v1'].apply(lambda x: 1 if x== 'spam' else 0)
```

```python
spam
```

```python
x_train , x_test, y_train, y_test = train_test_split(spam.v2, spam.spam)
```

```python
x_train
```

```python
x_test
```

```python
x_train.describe()
```

```python
cv = CountVectorizer()
x_train_count = cv.fit_transform(x_train.values)
```

```python
x_train_count.toarray()
```

```python
model= MultinomialNB()
model.fit(x_train_count,y_train)
```

```python
email_ham = ["hey wanna meet up for the game ?"]
email_ham_count= cv.transform(email_ham)
model.predict(email_ham_count)
```

```python
email_spam = ["reward money click "]
email_spam_count= cv.transform(email_spam)
model.predict(email_spam_count)
```

```python
x_test_count = cv.transform(x_test)
model.score(x_test_count, y_test)
```
